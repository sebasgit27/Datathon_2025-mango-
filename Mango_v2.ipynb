```python
import numpy as np

def var_score(y_true, y_pred):
    """
    Métrica VAR aproximada para evaluar modelos.

    y_true: producción real (lo que Mango produjo históricamente)
    y_pred: producción predicha por el modelo

    Idea:
    - Si produces menos que y_true → penaliza fuerte.
    - Si produces igual o más → bien, pero castiga un poco el exceso.
    """

    # Evitamos divisiones raras
    y_pred = np.maximum(y_pred, 1e-6)
    y_true = np.maximum(y_true, 1e-6)

    # Ratio base: si produces menos que lo "ideal", baja
    base_ratio = np.minimum(y_true / y_pred, 1.0)  # ≤ 1

    # Penalización suave por sobrestock (exceso de producción)
    overstock = np.maximum(y_pred - y_true, 0.0) / y_true  # porcentaje de exceso
    penalty = 0.1 * overstock  # el 0.1 es un peso que puedes ajustar

    score = base_ratio - penalty
    return score.mean()
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error
from catboost import CatBoostRegressor

# ============================================================
# 1. AGREGACIONES SEMANALES → POR PRODUCTO (ID)
# ============================================================

# Agregamos ventas y demanda por ID con más estadísticas
agg = df_train.groupby("ID").agg({
    "weekly_sales":  ["sum", "mean", "max", "min", "std"],
    "weekly_demand": ["sum", "mean", "max", "min", "std"]
})

# Limpiar nombres: ('weekly_sales', 'sum') -> 'weekly_sales_sum'
agg.columns = ["_".join(col) for col in agg.columns]
agg.reset_index(inplace=True);

# Unir agregados al train: 1 fila por ID con atributos + agregados
train_prod = df_train.drop_duplicates("ID").merge(agg, on="ID", how="left")

# Preparar test
test_prod = test.copy()

# Quitar columnas basura del CSV
TEST_EXCLUSIVOS = ['Unnamed: 28', 'Unnamed: 29', 'Unnamed: 30', 'Unnamed: 31', 'Unnamed: 32']
test_prod = test_prod.drop(columns=TEST_EXCLUSIVOS, errors='ignore')

# Unir agregados al test (cuando el ID ya estaba en train)
test_prod = test_prod.drop_duplicates("ID").merge(agg, on="ID", how="left")

# ============================================================
# 2. NUEVAS FEATURES DERIVADAS (ótiles para demanda/producción)
# ============================================================

# Ratio ventas/demanda: si está cerca de 1 o >1 indica posible falta de stock
train_prod["sales_demand_ratio"] = (
    train_prod["weekly_sales_sum"] / (train_prod["weekly_demand_sum"] + 1e-3)
)

# Volatilidad (desviación típica) de ventas y demanda
train_prod["sales_volatility"]  = train_prod["weekly_sales_std"].fillna(0)
train_prod["demand_volatility"] = train_prod["weekly_demand_std"].fillna(0)

# Igual para test (si las columnas existen tras el merge)
test_prod["sales_demand_ratio"] = (
    test_prod["weekly_sales_sum"] / (test_prod["weekly_demand_sum"] + 1e-3)
)
test_prod["sales_volatility"]  = test_prod["weekly_sales_std"].fillna(0)
test_prod["demand_volatility"] = test_prod["weekly_demand_std"].fillna(0)

# ============================================================
# 3. PREPARAR FEATURES Y TARGET
# ============================================================

target = "Production"

# Columnas que no queremos como features
drop_cols = ["weekly_sales", "weekly_demand", "num_week_iso", "year"]

train_prod = train_prod.drop(columns=drop_cols, errors="ignore")
test_prod  = test_prod.drop(columns=drop_cols, errors="ignore")

# Por seguridad, quitamos Production del test si apareciera
test_prod = test_prod.drop(columns=[target], errors="ignore")

# Lista de columnas de features (todas menos el target)
feature_cols = [c for c in train_prod.columns if c != target]

# Matrices finales
X_all = train_prod[feature_cols].copy()
y_all = train_prod[target].copy()
X_test = test_prod[feature_cols].copy()

# Detectar columnas categóricas y numéricas
cat_features = X_all.select_dtypes(include=object).columns.tolist()
num_features = X_all.select_dtypes(include=[np.number]).columns.tolist()

# ============================================================
# 4. FUNCIÓN PARA RELLENAR NaNs
# ============================================================

def fill_na(X, cat_features, num_features, ref=None):
    """
    Rellena NaNs en un DataFrame X:
    - categóricas con 'Unknown'
    - numéricas con la media segùn ref (si ref=None, usa X)
    Devuelve una copia rellenada.
    """
    X = X.copy()
    if ref is None:
        ref = X

    # Categóricas
    for col in cat_features:
        if col in X.columns:
            X[col] = X[col].fillna('Unknown')

    # Numéricas
    for col in num_features:
        if col in X.columns:
            m = ref[col].mean()
            X[col] = X[col].fillna(m)

    return X

# ============================================================
# 5. FUNCIÓN MÉTRICA VAR APROXIMADA (LOCAL)
# ============================================================

def var_score(y_true, y_pred, overstock_weight=0.1):
    """
    Métrica tipo VAR aproximada:
    - base_ratio = min(real / pred, 1.0)
      -> si produces menos que lo que deberías, baja.
      -> si produces igual o más, vale 1.
    - penalización suave por sobrestock (pred > real).
    """
    y_true = np.maximum(y_true, 1e-6)
    y_pred = np.maximum(y_pred, 1e-6)

    base_ratio = np.minimum(y_true / y_pred, 1.0)
    overstock = np.maximum(y_pred - y_true, 0.0) / y_true
    penalty = overstock_weight * overstock

    return (base_ratio - penalty).mean()

# ============================================================
# 6. VALIDACIÓN TEMPORAL POR TEMPORADA (NO ES PARA SUBMISSION)
# ============================================================

# Elegimos la óltima temporada disponible como validación (simula futuro)
last_season = train_prod["id_season"].max()
print("Temporada usada como validación:", last_season)

train_mask = train_prod["id_season"] != last_season
val_mask   = train_prod["id_season"] == last_season

X_tr = X_all.loc[train_mask].copy()
y_tr = y_all.loc[train_mask].copy()

X_val = X_all.loc[val_mask].copy()
y_val = y_all.loc[val_mask].copy()

# Rellenamos NaNs usando estadísticas de X_tr
X_tr_f  = fill_na(X_tr,  cat_features, num_features, ref=X_tr)
X_val_f = fill_na(X_val, cat_features, num_features, ref=X_tr)

# Entrenamos un modelo solo para evaluación (con early stopping)
model_eval = CatBoostRegressor(
    iterations=3000,
    learning_rate=0.04,
    depth=8,
    loss_function="RMSE",
    eval_metric="RMSE",
    random_seed=42,
    od_type="Iter",   # detector de overfitting por iteraciones
    od_wait=200,      # si en 200 iteraciones no mejora validación, para
    verbose=200
)

model_eval.fit(
    X_tr_f, y_tr,
    cat_features=cat_features,
    eval_set=(X_val_f, y_val),
    use_best_model=True,
    verbose=200
)

# Mejor número de árboles (iteraciones) encontrado por el early stopping
best_iter = model_eval.tree_count_
print("\nMejor número de iteraciones segùn validación temporal:", best_iter)

# Predicción en validación y métricas
val_pred = model_eval.predict(X_val_f)

rmse_val = np.sqrt(mean_squared_error(y_val, val_pred))
mae_val  = mean_absolute_error(y_val, val_pred)
var_val  = var_score(y_val.values, val_pred)

print("\n===== RESULTADOS VALIDACIÓN TEMPORAL ====")
print("RMSE validación:", rmse_val)
print("MAE  validación:", mae_val)
print("VAR simulado   :", var_val)
print("==========================================\n")

# ============================================================
# 7. BUSCAR MEJOR ALPHA (ESCALA) PARA MEJORAR VAR EN VALIDACIÓN
# ============================================================

alphas = [0.95, 1.00, 1.05, 1.10, 1.15, 1.20]
best_alpha = None
best_var = -1
best_rmse_alpha = None

print("Buscando mejor alpha para maximizar VAR simulado...\n")

for alpha in alphas:
    scaled_pred = val_pred * alpha
    var_alpha   = var_score(y_val.values, scaled_pred)
    rmse_alpha  = np.sqrt(mean_squared_error(y_val, scaled_pred))
    print(f"alpha={alpha:.2f} -> VAR={var_alpha:.6f}  |  RMSE={rmse_alpha:.2f}")

    if var_alpha > best_var:
        best_var = var_alpha
        best_alpha = alpha
        best_rmse_alpha = rmse_alpha

print("\n>>> Mejor alpha segùn VAR simulado:", best_alpha,
      "con VAR:", best_var, "y RMSE:", best_rmse_alpha)

# ============================================================
# 8. ENTRENAR MODELO FINAL CON TODO EL TRAIN Y GENERAR SUBMISSION
# ============================================================

# Rellenar NaNs en TODO el train y en el test usando estadísticas del train completo
X_all_f  = fill_na(X_all,  cat_features, num_features, ref=X_all)
X_test_f = fill_na(X_test, cat_features, num_features, ref=X_all)

# Entrenamos el modelo final usando best_iter
model_final = CatBoostRegressor(
    iterations=best_iter,    # usamos el nº de iteraciones óptimo encontrado
    learning_rate=0.04,
    depth=8,
    loss_function="RMSE",
    eval_metric="RMSE",
    random_seed=42,
    verbose=200
)

model_final.fit(
    X_all_f, y_all,
    cat_features=cat_features,
    verbose=200
)

# Predicción base de Production en el test
test_pred_raw = model_final.predict(X_test_f)

# Aplicamos el mejor alpha encontrado para favorecer la métrica VAR
production_test = test_pred_raw * best_alpha

# (Opcional: asegurar no negativos)
production_test = np.maximum(production_test, 0)

# Crear submission
submission = pd.DataFrame({
    "ID": test_prod["ID"],
    "Production": production_test
})

submission.to_csv("submission.csv", index=False)
print("\nArchivo 'submission.csv' generado.")
```

```python
# ============================================
#  Mango Datathon – LGBM con ventas/demanda agregadas
#  - Usa Production como target
#  - Usa weekly_sales y weekly_demand agregadas por ID
# ============================================

import pandas as pd
import numpy as np
from pathlib import Path

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

import lightgbm as lgb

from google.colab import drive
drive.mount('/content/drive')

DATA_DIR = Path("/content/drive/MyDrive/Datathon")


# -------------------------------------------------------------------
# Cargar datos + renombrar columnas Unnamed del test
# -------------------------------------------------------------------
def load_data():
    train_path = DATA_DIR / "train.csv"
    test_path = DATA_DIR / "test.csv"
    sample_sub_path = DATA_DIR / "sample_submission.csv"

    print("Leyendo CSV...")
    train = pd.read_csv(train_path, sep=';')
    test = pd.read_csv(test_path, sep=';')
    sample_sub = pd.read_csv(sample_sub_path)

    # Renombrar columnas Unnamed del test para que coincidan con train
    unnamed_test_cols = [c for c in test.columns if c.startswith("Unnamed")]
    if unnamed_test_cols:
        n_unnamed = len(unnamed_test_cols)
        last_train_cols = train.columns[-n_unnamed:]
        rename_map = {old: new for old, new in zip(unnamed_test_cols, last_train_cols)}
        print("Renombrando columnas Unnamed en test.csv:")
        print(rename_map)
        test = test.rename(columns=rename_map)

    print("train shape:", train.shape)
    print("test shape:", test.shape)
    print("sample_submission shape:", sample_sub.shape)

    return train, test, sample_sub


from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd

def parse_embedding(df, column_name='image_embedding'):
    """Convierte la columna de embedding (string) en un DataFrame de floats."""

    # 1. Asegurar que la columna es string y rellenar NaN si existen
    # (Aunque no deberían si ya limpiaste categóricas, es un paso de seguridad)
    df[column_name] = df[column_name].astype(str).str.replace('[', '').str.replace(']', '')

    # 2. Dividir la string por coma y crear un nuevo DataFrame
    embedding_df = df[column_name].str.split(',', expand=True).astype(float)

    # 3. Asignar nombres a las columnas (ej. embed_0, embed_1, ...)
    num_components = embedding_df.shape[1]
    embedding_df.columns = [f'embed_{i}' for i in range(num_components)]

    # 4. Reemplazar NaN (si la string estaba vacía) con 0
    embedding_df = embedding_df.fillna(0)

    return embedding_df

# -------------------------------------------------------------------
# Preparar features agregadas por ID
# -------------------------------------------------------------------
def prepare_features(train: pd.DataFrame, test: pd.DataFrame, sample_sub: pd.DataFrame):
    if "Production" not in train.columns:
        raise ValueError("Falta la columna 'Production' en train.csv")

    # 1) Agregar info semanal por ID en el TRAIN
    agg = train.groupby("ID").agg(
        weekly_sales_sum=("weekly_sales", "sum"),
        weekly_sales_mean=("weekly_sales", "mean"),
        weekly_sales_max=("weekly_sales", "max"),
        weekly_demand_sum=("weekly_demand", "sum"),
        weekly_demand_mean=("weekly_demand", "mean"),
        weekly_demand_max=("weekly_demand", "max"),
    ).reset_index()

    # 2) Crear dataset a nivel producto en TRAIN (una fila por ID)
    train_prod = train.drop_duplicates("ID").merge(agg, on="ID", how="left")

    # 3) Crear dataset a nivel producto en TEST alineado con sample_submission
    #    (para que el orden de X_test coincida con el orden de IDs del submit)
    test_uniq = test.drop_duplicates("ID")
    test_prod = sample_sub[["ID"]].merge(test_uniq, on="ID", how="left")
    test_prod = test_prod.merge(agg, on="ID", how="left")  # puede dar NaN si ID nuevo

    # 4) Definir y, X, X_test
    y = train_prod["Production"].astype(float)




    # =========================================================
    # PASO 1: PARSEAR EL EMBEDDING DE STRING A MATRIZ NUMÉRICA
    # =========================================================

    # Obtener la matriz numérica de embeddings solo para el set de entrenamiento
    X_embed_train = parse_embedding(train_prod.copy(), 'image_embedding')
    X_embed_test = parse_embedding(test_prod.copy(), 'image_embedding') # Necesario para aplicar el mismo PCA después

    scaler = StandardScaler()
    X_embed_train_scaled = scaler.fit_transform(X_embed_train)
    X_embed_test_scaled = scaler.transform(X_embed_test)

        # Definición del número óptimo de componentes
    K_OPTIMO = 119

    # 1. Re-inicializar PCA con K=119
    pca_final = PCA(n_components=K_OPTIMO, random_state=42)

    # 2. Ajustar y transformar el set de entrenamiento
    # Nota: X_embed_train_scaled debe ser la matriz estandarizada
    X_pca_train = pca_final.fit_transform(X_embed_train_scaled)

    # 3. Transformar el set de prueba usando el mismo ajuste del entrenamiento
    X_pca_test = pca_final.transform(X_embed_test_scaled)

    # 4. Crear los DataFrames de Componentes
    componentes_retenidas = [f'PC{i}' for i in range(1, K_OPTIMO + 1)]

    df_pca_train = pd.DataFrame(data=X_pca_train, columns=componentes_retenidas, index=train_prod.index)
    df_pca_test = pd.DataFrame(data=X_pca_test, columns=componentes_retenidas, index=test_prod.index)

    # Unir las 119 componentes principales al DataFrame de entrenamiento
    train_prod = pd.concat([train_prod, df_pca_train], axis=1)

    # Unir las 119 componentes principales al DataFrame de prueba
    test_prod = pd.concat([test_prod, df_pca_test], axis=1)

    print(f"✅ Integración PCA completa.")
    print(f"Dimensiones de df_train_x después de la unión: {train_prod.shape}")
    print(f"Dimensiones de df_test_x después de la unión: {test_prod.shape}")

    drop_cols = [
        "Production",
        "weekly_sales",
        "weekly_demand",
        "num_week_iso",
        "year",
    ]
    # también podemos quitar columns temporales de fechas si molestan, opcional

    # También quitamos 'image_embedding' por velocidad (si quieres luego lo metemos con PCA)
    if "image_embedding" in train_prod.columns:
        drop_cols.append("image_embedding")

    X = train_prod.drop(columns=[c for c in drop_cols if c in train_prod.columns])
    X_test = test_prod.drop(columns=[c for c in drop_cols if c in test_prod.columns])

    # Aseguramos mismas columnas en train y test (por si falta alguna en test)
    X_test = X_test[X.columns]

    # Detectar categóricas
    cat_cols = [c for c in X.columns if X[c].dtype == "object"]

    # Convertir a categoría para LGBM
    for c in cat_cols:
        X[c] = X[c].astype("category")
        X_test[c] = X_test[c].astype("category")

    print("Número de features:", X.shape[1])
    print("Categóricas:", len(cat_cols))

    return X, y, X_test, cat_cols


# -------------------------------------------------------------------
# Entrenamiento + validación rápida
# -------------------------------------------------------------------
def train_and_evaluate_lgbm(X, y, cat_cols, test_size=0.2, random_state=42):
    print("\n[VALIDACIÓN] Haciendo train/valid split...")
    X_train, X_valid, y_train, y_valid = train_test_split(
        X, y, test_size=test_size, random_state=random_state
    )

    print("X_train:", X_train.shape, " X_valid:", X_valid.shape)

    lgb_train = lgb.Dataset(X_train, label=y_train, categorical_feature=cat_cols, free_raw_data=False)
    lgb_valid = lgb.Dataset(X_valid, label=y_valid, categorical_feature=cat_cols, free_raw_data=False)

    params = {
        "objective": "regression",
        "metric": "rmse",
        "learning_rate": 0.05,
        "num_leaves": 31,
        "feature_fraction": 0.8,
        "bagging_fraction": 0.8,
        "bagging_freq": 1,
        "verbose": -1,
        "seed": 42,
    }

    print("\n[VALIDACIÓN] Entrenando LightGBM...")
    model = lgb.train(
        params,
        lgb_train,
        valid_sets=[lgb_train, lgb_valid],
        valid_names=["train", "valid"],
        num_boost_round=800,
        callbacks=[lgb.early_stopping(100, verbose=100)],
    )

    print("\n[VALIDACIÓN] Mejor iteración:", model.best_iteration)
    y_pred = model.predict(X_valid, num_iteration=model.best_iteration)
    rmse = np.sqrt(mean_squared_error(y_valid, y_pred))
    print(f"[VALIDACIÓN] RMSE: {rmse:.4f}")

    return model


# -------------------------------------------------------------------
# Entrenar con todo el train y generar submission
# -------------------------------------------------------------------
def train_full_and_predict_lgbm(X, y, X_test, cat_cols, sample_sub, best_iteration):
    print("\n[FINAL] Entrenando modelo LightGBM con TODO el train...")

    lgb_full = lgb.Dataset(X, label=y, categorical_feature=cat_cols, free_raw_data=False)

    params = {
        "objective": "regression",
        "metric": "rmse",
        "learning_rate": 0.05,
        "num_leaves": 31,
        "feature_fraction": 0.8,
        "bagging_fraction": 0.8,
        "bagging_freq": 1,
        "verbose": -1,
        "seed": 42,
    }

    model_full = lgb.train(
        params,
        lgb_full,
        num_boost_round=best_iteration
    )

    print("[FINAL] Prediciendo sobre test...")
    preds = model_full.predict(X_test, num_iteration=best_iteration)
    preds = np.maximum(preds, 0)  # evitar negativos

    submission = sample_sub.copy()
    submission["Production"] = preds

    return submission[["ID", "Production"]]


# -------------------------------------------------------------------
# Ejecutar
# -------------------------------------------------------------------
train, test, sample_sub = load_data()

X, y, X_test, cat_cols = prepare_features(train, test, sample_sub)

print("\n==== VALIDACIÓN RÁPIDA CON LGBM ====")
model_val = train_and_evaluate_lgbm(X, y, cat_cols)

best_iter = model_val.best_iteration

print("\n==== ENTRENAMIENTO FINAL Y SUBMISSION ====")
submission = train_full_and_predict_lgbm(X, y, X_test, cat_cols, sample_sub, best_iter)

out_path = DATA_DIR / "submission_lgbm_agg.csv"
submission.to_csv(out_path, index=False)
print(f"\n✅ Submission guardada en:\n{out_path}")
```
